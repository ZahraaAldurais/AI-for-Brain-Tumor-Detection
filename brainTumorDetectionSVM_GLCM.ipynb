```python
import numpy as np
import cv2
import os
from skimage.feature import greycomatrix, greycoprops
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
yes_dir = os.path.join(data_dir, "yes")
yes_images = []
yes_labels = []
for img_name in os.listdir(yes_dir):
  img = cv2.imread(os.path.join(yes_dir, img_name))
  yes_images.append(img)
  yes_labels.append(1) # 1 for tumor
```

```python
no_dir = os.path.join(data_dir, "no")
no_images = []
no_labels = []
for img_name in os.listdir(no_dir):
  img = cv2.imread(os.path.join(no_dir, img_name))
  no_images.append(img)
  no_labels.append(0) # 0 for normal
```

```python
images = yes_images + no_images
labels = yes_labels + no_labels
```

```python
glcm_features = []
for image in images:
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  glcm = greycomatrix(gray, [1], [0, np.pi/4, np.pi/2, 3*np.pi/4], levels=256, normed=True, symmetric=True)
  glcm_features.append(np.concatenate([greycoprops(glcm, prop).ravel() for prop in ['contrast', 'energy', 'homogeneity', 'correlation']]))
```

```python
X_train, X_test, y_train, y_test = train_test_split(glcm_features, labels, test_size=0.2, random_state=42)

clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```
